```sage
import socket

def sendline(sock, data):
    sock.sendall(data + b"\n")

def recvline(sock):
    buf = b""
    while True:
        data = sock.recv(1)
        if not data:
            break
        buf += data
        if data == b"\n":
            break
    return buf

def recvuntil(sock, delim):
    buf = b""
    while True:
        data = sock.recv(1)
        if not data:
            break
        buf += data
        if delim in buf:
            break
    return buf
```

```sage
import gmpy2

def fermat_factorization(n):
    factor_list = []
    #gmpy2.get_context().precision = 2048
    a = int(gmpy2.sqrt(n))

    a2 = a * a
    b2 = gmpy2.sub(a2, n)

    while True:
        a += 1
        b2 = a * a - n

        if gmpy2.is_square(b2):
            b2 = gmpy2.mpz(b2)
            gmpy2.get_context().precision = 2048
            b = int(gmpy2.sqrt(b2))
            factor_list.append([a + b, a - b])

        if len(factor_list) == 2:
            break

    return factor_list

fermat_factorization(n)
```

```sage
num_variables = n  # You can change this to any number of variables

# Define the polynomial ring with the number of variables
R = PolynomialRing(Integers(p), num_variables, names='x')

# Define the variables
variables = R.gens()

# Define the equations
equations = []
for i in range(1, n):  # Include up to n
    equations.append(sum([v^i for v in variables]))

# Compute the Groebner basis
I = R.ideal(equations)
GB = I.groebner_basis()

# Convert to polynomial system in Z/pZ
Pol_p = PolynomialRing(ZZ(p), num_variables, names='x')
GB_mod_p = [f.change_ring(ZZ(p)) for f in GB]

# Solve the system modulo p
solutions = Pol_p.quotient(GB_mod_p)

# Print solutions
print(solutions)
```
